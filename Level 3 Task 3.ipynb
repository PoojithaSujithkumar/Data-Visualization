```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Sujithkumar/Desktop/Cognifyz/Dataset .csv")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'].dropna(), bins=20, kde=True, color='teal')
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Aggregate rating', palette='Set2')
plt.title("Count of Unique Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
df['Primary Cuisine'] = df['Cuisines'].astype(str).apply(lambda x: x.split(',')[0].strip())
```

```python
top_cuisines = df['Primary Cuisine'].value_counts().head(10).index
df_top_cuisines = df[df['Primary Cuisine'].isin(top_cuisines)]

plt.figure(figsize=(10,5))
sns.barplot(data=df_top_cuisines, x='Primary Cuisine', y='Aggregate rating', palette='coolwarm')
plt.title("Average Rating of Top 10 Cuisines")
plt.xticks(rotation=45)
plt.ylabel("Average Rating")
plt.xlabel("Cuisine")
plt.show()
```

```python
top_cities = df['City'].value_counts().head(10).index
df_top_cities = df[df['City'].isin(top_cities)]

plt.figure(figsize=(10,5))
sns.barplot(data=df_top_cities, x='City', y='Aggregate rating', palette='crest')
plt.title("Average Rating by Top 10 Cities")
plt.xticks(rotation=45)
plt.ylabel("Average Rating")
plt.xlabel("City")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Votes', y='Aggregate rating', alpha=0.5)
plt.title("Votes vs. Aggregate Rating")
plt.xlabel("Votes")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(data=df, x='Price range', y='Aggregate rating', palette='pastel')
plt.title("Aggregate Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
if 'Has Online delivery' in df.columns:
    plt.figure(figsize=(7,4))
    sns.boxplot(data=df, x='Has Online delivery', y='Aggregate rating')
    plt.title("Aggregate Rating: Online Delivery vs. No Delivery")
    plt.show()
```

```python
df['Online_Delivery_Flag'] = df['Has Online delivery'].apply(lambda x: 1 if str(x).lower() == 'yes' else 0)
df['Table_Booking_Flag'] = df['Has Table booking'].apply(lambda x: 1 if str(x).lower() == 'yes' else 0)

plt.figure(figsize=(10,6))
sns.heatmap(df[['Aggregate rating', 'Votes', 'Price range', 'Online_Delivery_Flag', 'Table_Booking_Flag']].corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Heatmap")
plt.show()
```
